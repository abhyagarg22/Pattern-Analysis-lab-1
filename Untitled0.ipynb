```python
import torch
import matplotlib.pyplot as plt

# Select device (GPU if available, else CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Create coordinate grids using torch
x_vals = torch.arange(-4.0, 4.0, 0.01, device=device)
y_vals = torch.arange(-4.0, 4.0, 0.01, device=device)
X, Y = torch.meshgrid(x_vals, y_vals, indexing='ij')

# Calculate the 2D Gaussian function
Z = torch.exp(-(X**2 + Y**2) / 2.0)

# Move to CPU for plotting
Z_cpu = Z.cpu().numpy()

# Plot the Gaussian
plt.imshow(Z_cpu, cmap='viridis')
plt.title('2D Gaussian Function (PyTorch)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Choose device: GPU if available, otherwise CPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Create meshgrid coordinates
X, Y = np.mgrid[-4.0:4.0:0.01, -4.0:4.0:0.01]

# Convert to PyTorch tensors
x = torch.tensor(X, dtype=torch.float32, device=device)
y = torch.tensor(Y, dtype=torch.float32, device=device)

# 2D sine function: angle depends on both x and y
# You can tweak frequency_x and frequency_y for more or fewer stripes
frequency_x = 2.0
frequency_y = 0.0
z = torch.sin(frequency_x * x + frequency_y * y)

# Move data back to CPU and plot
plt.imshow(z.cpu().numpy(), cmap='grey', origin='lower')
plt.colorbar(label='Sine value')
plt.title("2D Sine Function Stripes")
plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Select device (GPU if available)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Create meshgrid for coordinates
X, Y = np.mgrid[-4.0:4.0:0.01, -4.0:4.0:0.01]
x = torch.tensor(X, dtype=torch.float32, device=device)
y = torch.tensor(Y, dtype=torch.float32, device=device)

# ----- Step 1: Create Gaussian -----
gaussian = torch.exp(-(x**2 + y**2) / 2.0)

# ----- Step 2: Create 2D sine function -----
frequency_x = 3.0  # change to adjust stripe frequency in x
frequency_y = 0.0  # set to nonzero for tilted stripes
sine_func = torch.sin(frequency_x * x + frequency_y * y)

# ----- Step 3: Multiply Gaussian and sine (modulation) -----
gabor = gaussian * sine_func

# Gabor
plt.imshow(gabor.cpu().numpy(), cmap='seismic', origin='lower')
plt.title("Gabor Filter (Gaussian  Sine)")
plt.colorbar()

plt.tight_layout()
plt.show()

```
