```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Select device (GPU if available)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Create meshgrid for coordinates
X, Y = np.mgrid[-4.0:4.0:0.01, -4.0:4.0:0.01]
x = torch.tensor(X, dtype=torch.float32, device=device)
y = torch.tensor(Y, dtype=torch.float32, device=device)

# ----- Step 1: Create Gaussian -----
gaussian = torch.exp(-(x**2 + y**2) / 2.0)

# ----- Step 2: Create 2D sine function -----
frequency_x = 3.0  # change to adjust stripe frequency in x
frequency_y = 0.0  # set to nonzero for tilted stripes
sine_func = torch.sin(frequency_x * x + frequency_y * y)

# ----- Step 3: Multiply Gaussian and sine (modulation) -----
gabor = gaussian * sine_func

# Gabor
plt.imshow(gabor.cpu().numpy(), cmap='seismic', origin='lower')
plt.title("Gabor Filter (Gaussian  Sine)")
plt.colorbar()

plt.tight_layout()
plt.show()

```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Device configuration (GPU if available)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Using device:", device)

# Define grid in the complex plane
# Adjust spacing for resolution (smaller = higher resolution, slower)
Y, X = np.mgrid[-1.3:1.3:0.005, -2:1:0.005]

# Convert to PyTorch tensors
x = torch.tensor(X, dtype=torch.float32, device=device)
y = torch.tensor(Y, dtype=torch.float32, device=device)
z = torch.complex(x, y)    # The constant c
zs = z.clone()             # Current iteration values
ns = torch.zeros_like(z, dtype=torch.float32, device=device)  # Divergence counter

# Mandelbrot iteration
max_iter = 200
for i in range(max_iter):
    zs_ = zs * zs + z
    not_diverged = torch.abs(zs_) < 4.0
    ns += not_diverged
    zs = zs_

# Function to process iteration counts into colors
def processFractal(a):
    """Convert iteration counts into an RGB fractal image."""
    a_cyclic = (6.28 * a / 20.0).reshape(list(a.shape) + [1])
    img = np.concatenate([
        10 + 20 * np.cos(a_cyclic),
        30 + 50 * np.sin(a_cyclic),
        155 - 80 * np.cos(a_cyclic)
    ], axis=2)
    img[a == a.max()] = 0
    return np.uint8(np.clip(img, 0, 255))

# Convert back to NumPy for plotting
ns_cpu = ns.cpu().numpy()

# Plot result
plt.figure(figsize=(16, 10))
plt.imshow(processFractal(ns_cpu))
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
